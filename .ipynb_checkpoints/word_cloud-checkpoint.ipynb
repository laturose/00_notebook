```python
import random
p1=["内部数据","交易授权","航空订单","IVR进线","资料异动","开卡数据","现金分期","资格查询","试算申请","查询密码变更校验","三星PAY数据","华为支付数据","快捷支付管理数据","手机银行","PAD银行","网上银行登录数据","备用金","预借现金","交易密码变更校验","动码验证结果","移动定位和运营商客户数据","境内外商户位置数据","客户进线行程报备","模型评分数据","网上商城订单数据","注册","登录","绑卡","支付","修改密码","免密限额","理财","消费金融","E招贷","HEC数据","Apple Pay数据","临额申请","支付功能设置","交易密码","电子行用卡","电子信用卡","修改手机号"]
p2=["公安部黑名单","人行黑名单","代理IP评分","电信数据","联通数据","运营商客户数据","运营商数据","移动定位","标准类数据","黑灰名单","黑IP","黑MAC","黑手机号","代理服务器名单","各类风险通报","社工库","IP地址解析","GPS定位","试金石"]
color_list=['484242','808080','C0C0C0']



for tg_list in p1,p2:
    randlist = [random.randint(1,len(tg_list)-1) for _ in range(4*len(tg_list))]
    p1_adj = [tg_list[i] for i in randlist]
    cnt = 0
    for wd in p1_adj:
        cnt = cnt +1
        weight = random.randint(1,1)
        color = random.randint(1,3)
        color = color_list[color-1]
        line = "{word};{weight};{color};{angle};{font};{repeat};{url}".format(
        word=wd,
        weight=weight,
            color=color,
            angle="",
            font="",
            repeat="",
            url=""
        )
        print(line)
    print("*"*100)
    print("*"*100)
    print("*"*100)
```
