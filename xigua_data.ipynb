```python


import requests
import re
import json
import hashlib
```

```python
mode = 'p'

def genCheckSum(body, secret):
    encryStr = body + secret;
    m = hashlib.md5(encryStr.encode("utf-8"))
    encryStrMd5 = m.hexdigest().lower()
    return encryStrMd5[14:18]

secret = 'ded15192-6275-4cc4-be22-d2c2f5a7adee'
userid = 21012

# POST http://openapi.xiguaji.com/subsbiz/AddLevelHighPriority  HTTP/1.1

# HttpRequest-Header
# Content-Type: application/json
# Host: openapi.xiguaji.com
# userid: 21012
# checksum: 773

# HttpRquest-Body
sub_data = {
    "WechatId": "xiguaxiaojun",
    "PriorityLevel":10
}

his_data = {"wechatid":"xiguaxiaojun","DatgeRangeType":1}


pull_data = { "LastSubscriptionID": 0}
#data = {"WechatId": "iloveajmd","DatgeRangeType": 1 }


his_api_url = 'http://openapi.xiguaji.com/v3/Article/GetHistoryArticleByDateRange'
subsbiz_api_url = 'http://openapi.xiguaji.com/v3/subsbiz/AddLevelHighPriority'
pull_api_url = 'http://openapi.xiguaji.com/v3/Article/PullIncrement'

if mode == 'sub':
    data = sub_data
    url = subsbiz_api_url
elif mode == 'his':
    data = his_data
    url = his_api_url
elif mode == 'p':
    data = pull_data
    url = pull_api_url

dataJson = json.dumps(data)


checksum = genCheckSum(dataJson, secret)

headers = {"Content-Type":"application/json",  "userid":str(userid), "checksum":checksum}

r = requests.post(url=url, headers=headers, data = dataJson )
#t = json.loads(r.text)
t = json.loads(r.content.decode("utf8"))
```
